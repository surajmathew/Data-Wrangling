### Step 1 - Importing the necessary libraries

```python
import matplotlib.pyplot as plt
import random                    # Random number generation
from faker import Faker          # Mock data generator
fake = Faker('en_IN')            # Set locality to India 
import colorama                  # For font styling and color of the output.
from colorama import Fore, Style
```

### Step 2 - Randomly generating 12 names with their heights in cm and weights in kg

```python
lv_names = []         #Empty list to hold the randomly generated names
lv_heights_cm = []    #Empty list to hold the randomly generated heights
lv_weights_kg = []    #Empty list to hold the randomly generated weights

num_elements = 12 # The desired number of elements in each list. 
Faker.seed(12)    # Remove or change the number for a new set of friends.
random.seed(12)   # Remove or change the number for new BMIs.

for _ in range(num_elements): 
    lv_names.append(fake.name())
    lv_heights_cm.append(random.randint(145, 190))
    lv_weights_kg.append(random.randint(50, 120))

print(Style.BRIGHT + '\nNAME\t\t\tHEIGHT\tWEIGHT' + Style.RESET_ALL)
for i in range(len(lv_names)):
    # If statement to acheive the same space between name and heights.
    if len(lv_names[i]) <= 15: # Checks the length of the name.
        print(f'{lv_names[i]}\t\t{lv_heights_cm[i]}cm\t{lv_weights_kg[i]}kg')
    else:
        print(f'{lv_names[i]}\t{lv_heights_cm[i]}cm\t{lv_weights_kg[i]}kg')   
```

<p style="color:#204060; font-size:15px" >
First we start by declaring three empty lists. Then we set the desired number of elements to be generated (in this case 12) for each list with <i>"num_elements"</i>. After than, we set the seeds for the random name, weight and height generation. The seeds allow us to reproduce the same data (freeze the data once generated) each time we run the code. 
</p>

<p style="color:#204060; font-size:15px" >
After that we use a for loop in the range of the <i>"num_elements"</i> to append to each list the 12 random data. The attributes given in the <i>"rand_int()"</i> function represent the ranges within which the random numbers need to be generated. Finally, we print the contents of our list again using a for loop. To ensure that the displayed results are in the proper format, we use an if-else statement equally space the elements 
</p>

### Step 3 - Calculating the BMI using the BMI formula and iterating the 12 names through a for loop

```python
# Code to calculate the Original BMI for every element in the list
bmis = [] # New empty list for the BMI's

for i in range(len(lv_heights_cm)): #iterating through the elements in the name list using a counter and calulating the BMI
    bmis.append(lv_weights_kg[i]/((lv_heights_cm[i]/100)**2))#BMI Formula and appending each of the calculated BMI into the bmi empty list

#Formatting the Output
print(Style.BRIGHT + 'BMIs:' + Style.RESET_ALL)
for bmi in bmis:
    print(' ', bmi)
```

<p style="color:#20568B; font-size:15px" >
We start by delcaring our list <i>"bmis"</i> After that we use a for loop to calculate each of the BMIs using the BMI formula and append each result to the list <i>"bmis"</i>. We then print the contents of the list with another for loop.</p><br>

### Appending the BMIs calculated to the existing Table

```python
print(Style.BRIGHT + '\nNAME\t\t\tHEIGHT\tWEIGHT\tBMI' + Style.RESET_ALL) #Setting the headers and its formatting

for i in range(len(lv_names)):
    # The below If statement is to acheive the same space between name and heights.
    if len(lv_names[i]) <= 15: # This is to check the length of the name.
        print(f'{lv_names[i]}\t\t{lv_heights_cm[i]}cm\t{lv_weights_kg[i]}kg\t{bmis[i]:.2f}')
    else:
        print(f'{lv_names[i]}\t{lv_heights_cm[i]}cm\t{lv_weights_kg[i]}kg\t{bmis[i]:.2f}')
```

### Step 4 - Calculating the NEW BMI using the NEW BMI formula and iterating the 12 names through a for loop

```python
# Code to calculate the NEW BMI for every element in the list
new_bmis = [] # New empty list for the NEW BMI's

for j in range(len(lv_heights_cm)): #iterating through the elements in the name list using a counter and calulating the BMI
    new_bmis.append(1.3*(lv_weights_kg[j] / ((lv_heights_cm[j]/100) ** 2.5)))#NEW BMI Formula and appending each of the calculated BMI into the bmi empty list

#Formatting the Output
print(Style.BRIGHT + 'NEW BMIs:' + Style.RESET_ALL)
for new_bmi in new_bmis:
    print(' ', new_bmi)
```

<p style="color:#20568B; font-size:15px" >
We start by delcaring our list <i>" new_bmis"</i> before using a for loop to calculate each of the NEW BMIs and appending each result to the list <i>"new_bmis"</i>. We then print the contents of the list with another loop.</p><br>

### Appending the NEW BMIs calculated to the existing Table

```python
print(Style.BRIGHT + '\nNAME\t\t\tHEIGHT\tWEIGHT\tBMI\tNEWBMI' + Style.RESET_ALL)

for k in range(len(lv_names)):
    # The below If statement is to acheive the same space between name and heights.
    if len(lv_names[k]) <= 15: # This is to check the length of the name.
        print(f'{lv_names[k]}\t\t{lv_heights_cm[k]}cm\t{lv_weights_kg[k]}kg\t{bmis[k]:.2f}\t{new_bmis[k]:.2f}')
    else:
        print(f'{lv_names[k]}\t{lv_heights_cm[k]}cm\t{lv_weights_kg[k]}kg\t{bmis[k]:.2f}\t{new_bmis[k]:.2f}')
```

```python
lv_colours = []

for bmi in bmis:
    if bmi < 18.5:
        lv_colours.append('c')
    elif bmi < 25:
        lv_colours.append('g')
    elif bmi < 30:
        lv_colours.append('m')
    else:
        lv_colours.append('r')
```

The above piece of code first declares an empty list to hold the colours defined based on the conditions. The condition are mentioned in the following if-else statements inside a for loop. Each of the

<p style="color:#20568B; font-size:15px" >
The funtion below <i>"bmi_results"</i> is a custom function written that takes a BMI as an argument and returns the corresponding results that satifies the condition and colours the output text with the same colour defined as part of lv_colours
</p>

```python
def bmi_results(bmi):
    if bmi < 18.5:
        print(Fore.CYAN, end='')
        return 'which is underweight'
    elif bmi < 25:
        print(Fore.GREEN, end='')
        return 'which is normal weight'
    elif bmi < 30:
        print(Fore.MAGENTA, end='')
        return 'which is overweight'
    else:
        print(Fore.RED, end='')
        return 'which is obese'
```

```python
print('Legend:', end='')
print(Fore.CYAN + ' Underweight ' + Fore.GREEN + ' Normal Weight ' + Fore.MAGENTA +
     ' Overweight ' + Fore.RED + ' Obese\n' + Fore.BLACK)
    
for i in range(len(lv_names)): # For each name, print their BMI results.
    print(Fore.BLACK, end='') # Resets the colour at each loop
# If statement to acheive the same space between name and sentence. 
    if len(lv_names[i]) <= 15:
        print(f'{lv_names[i]}\t\thas a BMI of {bmis[i]:.2f} {bmi_results(bmis[i])}.The NEW BMI index is {new_bmis[i]:.2f}')
    else:
        print(f'{lv_names[i]}\thas a BMI of {bmis[i]:.2f} {bmi_results(bmis[i])}.The NEW BMI index is {new_bmis[i]:.2f}')
```

<p style="color:#20568B; font-size:15px" >
The above code, is written to print the names of the people, their calculated BMI and the NEW BMI in the colour of the BMI category and the result. Formatting is done to display the results in an ordered fashion

**Using the ZIP Function**

```python
bmi_categories =[]
for name, height, weight in zip(lv_names,lv_heights_cm,lv_weights_kg):
      bmi = weight/ ((height/100) ** 2) #BMI Formula
      bmi_new = 1.3*(weight / ((height/100) ** 2.5)) #NEW BMI Formula
      if bmi < 18.5:
        bmi_category = "underweight"
        bmi_categories.append(bmi_category)
      elif bmi < 25.0:
        bmi_category = "normal"
        bmi_categories.append(bmi_category)
      elif bmi < 30.0:
        bmi_category = "overweight"
        bmi_categories.append(bmi_category)
      else:
        bmi_category = "obese"
        bmi_categories.append(bmi_category)
      print(f"{name}has BMI of {bmi:.2f}, which is {bmi_category} . The new BMI index is {bmi_new:.2f}")
```

<h3 style="color:#336699">Body Mass Index</h3>

<p style="color:#20568B; font-size:15px";>
    Body Mass Index (BMI) is a simple measurement relative to one's height and weight and it is calculated by dividing the weight in kilograms by the height in meters squared; <i>weight (kg) &#8725; height (m) <sup>2</sup></i>. The number resulting out of this formula is the BMI measure. BMI indicates an indiviual's weight category based on ranges instituted by the World Health Organisation. Since this is a ratio, it is a basic indication of one's health and not any other information.Although BMIs have been defined by WHO as statistical categories, there are slight variations in the ranges that differ from country to country based on the decents and historic lineage of the people in a particular region.For example in HongKong the normal range is between 18.5 and 22.9 as opposed to the WHO defined range of 18.5 and 24.9. Therefore based on the decent and the region these ranges and categorizations can vary for an Indian vs a Britisher Vs an American.
</p>
<p style="color:#20568B; font-size:15px";>
    The intended purpose of BMIs is to measure the rates of overweight and obesity in a population. It helps giving us insights into potential health risks and disease at large. This will inturn help in planning and allocating resouces, programs and government health schemes that will help imporve the lifestyles of its people. Medical professionals use BMI as more of a guide to help investigate high-risk health issues associated with a particular BMI range.
</p>
<p style="color:#20568B; font-size:15px";>
    As mentioned, a BMI is nothing more than an indicator. There can be cases where the BMI would be really high but the individual turns out to be healthy. For example based on the sport, athletes can be categorised as underweight or obese (but has complete muscle mass). BMI therefore, should be used in conjunction with other health parameters to generate a complete picture of one's individual health. There are a lot of limitations to these catetorizations as well, such as scaling (as the weight and height are not in the same dimension), physical characteristics of an individual, societal way of life, muscle mass vs fat. The NEW BMI is another measure of corpulance index where an exponent of 2.5 is used in the formula. This minimizes the distortions for shorter and taller individuals that was found using the normal BMI formula with an exponent of 2. A multiplier of 1.3 was used to bring the new BMI values close to the BMI calucated figures. It fits more closed on the empirical data with lesser distoration as compared to an exponent 2 or exponent 3.

</p>
<p style="color:#20568B; font-size:15px";>
     The ultimate aim is to fight eating disorders and minimize junk food intake. The insurance industry plays on the BMI component extensively. Recently I got myself a health insurance for my entire family. Since my BMI is on the higher side, I had to pay extra premium for it. I was classified as a "high risk BMI applicant", though my other health parameters are all normal.
</p>
<p style="color:#20568B; font-size:15px";>
    In conclusion, BMI is a good general health indicator. But it should never be seen in isolation as an indicator of one's health. We need to see BMI in conjunction with other health tests and parameters to understand the health condition of a person. Only then we should take corrective and prevention action, measures or treatment.
